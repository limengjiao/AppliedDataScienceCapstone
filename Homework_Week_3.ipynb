```python
#import relevant tools
import numpy as np
import pandas as pd 
from bs4 import BeautifulSoup
import urllib.request
import bs4 as bs
```

```python
# Read Wiki HTML
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
soup = bs.BeautifulSoup(source,'html.parser')
```

```python
# Read table by rows
table = soup.find('table')
table_rows = table.find_all('tr')
```

```python
# Retrive data in table
rows = []
columns_1 = []
columns_2 = []
columns_3 = []
for tr in table_rows:    
    tds = tr.find_all("td")
    s_1 = ''
    s_2 = ''
    s_3 = ''
    if len(tds) == 3:
        #get td[0] text
        s_1 = tds[0].get_text()
        s_1 = s_1.replace('\n', '')
        columns_1.append(s_1)
        
        #get td[1] text
        s_2 = tds[1].get_text()
        s_2 = s_2.replace('\n', '')
        columns_2.append(s_2)
        
        #get td[2] text
        s_3 = tds[2].get_text()
        s_3 = s_3.replace('\n', '')
        if s_3 == '':
            columns_3.append('null')
        else:
            s_3 = s_3.replace('/', ',')
            columns_3.append(s_3)       
    
# Put value to dataframe  
df = pd.DataFrame(rows, columns=['PostalCode','Borough','Neighborhood'])
df['PostalCode'] = columns_1
df['Borough'] = columns_2
df['Neighborhood'] = columns_3
df.head(10)
```

```python
# Drop Not assigned borough's rows
index = df[ df['Borough'] == 'Not assigned'].index
df.drop(index , inplace=True)
df.head(20)
```

```python
# Replace not assigned Neighborhoods with Borough if true
index_2 = df[df['Neighborhood'] == 'null'].index
#print(len(index_2))
for index in range(len(index_2)):    
    df['Neighborhood'].replace({'null': df.loc[index, 'Borough'].values[0]}, inplace=True)
```

```python
#Reset index
df.reset_index(inplace = True, drop = True)
df.head(10)
```

```python
#Merge all data to one dataframe
df_2 = pd.merge(left=df, right=df_gd, left_on='PostalCode', right_on='Postal Code')
```
