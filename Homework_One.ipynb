```python
#import relevant tools
import numpy as np
import pandas as pd 
from bs4 import BeautifulSoup
import urllib.request
import bs4 as bs
```

```python
# Read Wiki HTML
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
soup = bs.BeautifulSoup(source,'html.parser')
```

```python
# Read table by rows
table = soup.find('table')
table_rows = table.find_all('tr')
```

```python
# Retrive data in table
rows = []
columns_1 = []
columns_2 = []
columns_3 = []
for tr in table_rows:    
    tds = tr.find_all("td")  
    
    for td in tds:
        bs = td.find_all("b")
        for b in bs:
            columns_1.append(b.get_text()) 
            
        s = ""
        span = td.find("span")
        s = span.get_text()       
        if '(' in s:
            s_1 = s.split('(')[0].replace(" /", ",")
            s_2 = s.split('(')[-1].replace(" /", ",").replace(")", "")
        else:
            s_1 = s.replace(" /", ",")
            s_2 = 'null'
            
        columns_2.append(s_1)
        columns_3.append(s_2)
        
        
# Put value to dataframe  

df = pd.DataFrame(rows, columns=['PostalCode','Borough','Neighborhood'])
df['PostalCode'] = columns_1
df['Borough'] = columns_2
df['Neighborhood'] = columns_3
df.head(30)
```

```python
# Drop Not assigned borough's rows
index = df[ df['Borough'] == 'Not assigned'].index
df.drop(index , inplace=True)
df.head(20)
```

```python
# Replace not assigned Neighborhoods with Borough
index_2 = df[df['Neighborhood'] == 'null'].index
df['Neighborhood'].replace({'null': df.loc[index_2, 'Borough'].values[0]}, inplace=True)
```

```python
#Reset index
df.reset_index(inplace = True, drop = True)
df.head(30)
```
