```python
import pandas as pd
import numpy as np
```

```python
#Read Global Confirmd Data
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df = pd.read_csv(url)
```

```python
df.head(10)
```

```python
#Add All Region/Country Confirmed Cases as Global Total Confirmed 
df_g1 = df.drop(columns = ['Province/State', 'Country/Region', 'Lat', 'Long'])
df_g1.head(10)
```

```python
df_g1.loc['total'] = df_g1.select_dtypes('number').sum()
```

```python
df_g1 = df_g1.loc['total']
```

```python
df_g1
```

```python
df_g1.shape
```

```python
#Create a New DataFrame with Date and Global Total Confirmed Cases
df_g2 = pd.DataFrame()
```

```python
df_g2 = df_g2.append(df_g1, ignore_index = True)
```

```python
df_g2.shape
```

```python
df_g2.index = ['ConfirmNumer']
df_g2
```

```python
#Draw Global Confirmed Case Increase Trend Map
import matplotlib.pyplot as plt
```

```python
df_g2 = df_g2.transpose()
df_g2
```

```python
df_g2.reset_index(inplace=True)
df_g2.rename(columns={'index': 'Date'},inplace=True)
```

```python
from datetime import datetime, timedelta
df_g2['Date'] = pd.to_datetime(df_g2['Date'])
df_g2.sort_values(by=['Date'], inplace=True, ascending=True)
df_g2.reset_index(inplace=True)
```

```python
df_g2.plot.line(x='Date', y='ConfirmNumer', title='Global Confirm Cases', color='maroon', figsize=(15,5))
plt.xlabel('Date')
plt.ylabel('Confirm Number')
plt.show()
```

```python
#Get China Total Confirmed Cases
df_c1 = df.drop(columns = ['Province/State', 'Lat', 'Long'])
df_c2 = df_c1.loc[df_c1['Country/Region']== 'China']
df_c2.loc['Total'] = df_c2.select_dtypes('number').sum()
df_c3 = df_c2.loc['Total'] 
df_c3.head(30)
```

```python
df_China = pd.DataFrame()
df_China = df_China.append(df_c3, ignore_index = True)
df_China.index = ['China']
df_China= df_China.transpose()
df_China.reset_index(inplace=True)
df_China.rename(columns={'index': 'Date'},inplace=True)
df_China.dropna(axis=0, inplace=True)
df_China.tail(10)
```

```python
from datetime import datetime, timedelta
df_China['Date'] = pd.to_datetime(df_China['Date'])
df_China.sort_values(by=['Date'], inplace=True, ascending=True)
df_China.reset_index(inplace=True)
```

```python
df_China.tail(60)
```

```python
#Plot China Confirmed Cases Trend Map
import matplotlib.pyplot as plt
df_China.plot.line(x='Date', y='China', title='China Total Confirm Cases', color='OrangeRed', figsize=(15,5))
plt.xlabel('Date')
plt.ylabel('Confirm Number')
plt.show()
```
